```python
# 3. Create a list of 3 students, each represented as a dictionary:
# Each dictionary should have keys: "name", "age", "grade"
students = [
    {"name": "Alice", "age": 14, "grade": "B"},
    {"name": "Bob", "age": 15, "grade": "A"},
    {"name": "Charlie", "age": 16, "grade": "C"}
]
```

```python
# 3.1. Loop through the list and print the name and grade of each student.
for stdt in students:
  print(stdt["name"],":", stdt["grade"])
```

```python
# 3.2. Write a loop to calculate the average grade (assuming grades are numbers).
total_mark = 0
grade_scale = {"A":80, "B":70, "C": 60}
for stdt in students:
  total_mark += grade_scale[stdt["grade"]]
print("Average: ",total_mark/len(students))
```

```python
# 4. Create a dictionary to count how many times each word appears in this list:
# Output should be:
# {"apple": 3, "banana": 2, "cherry": 1}
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
count = {}
for word in words:
  if word in count:
    count[word] += 1
  else:
    count[word] = 1
print(count)
```

```python
# 5. Build a simple phonebook:
# Ask the user to enter names and phone numbers
# Store them in a dictionary
# Allow the user to look up numbers by name
print("Phonebook Menu\n")
print("1. Add a new contact\n")
print("2. Look up a contact\n")
print("3. Quit\n")

phonebook ={}

while True:
  try:
    print("You have ", len(phonebook), " contacts")
    choice = int(input("Enter your choice: "))
    if choice == 1:
      name = input("Enter name: ")
      phone = input("Enter phone number: ")
      phonebook[name] = phone
      print("Contact added successfully")
    elif choice == 2:
      name = input("Enter name: ")
      if name in phonebook:
        print(name,":",phonebook[name])
      else:
        print("Contact not found")
    elif choice == 3:
      print("Goodbye")
      break
  except ValueError:
    print("Invalid choice")
```

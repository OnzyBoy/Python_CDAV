```python
# 8. Ask for a grade (0–100) and print:
# "A" if 80 and above
# "B" if 60–79
# "C" if 40–59
# "F" if below 40

def grade_calculator(mark):
  if mark >= 80:
    return "A"
  elif mark >= 60:
    return "B"
  elif mark >= 40:
    return "C"
  else:
    return "F"


score = float(input("Enter your score to see your grade: "))
if 0 <= score <= 100:
  grade = grade_calculator(score)
  print(f"Your grade is {grade}")
else:
  print("Invalid input. Please enter a score between 0 and 100.")
```

```python
# 9. Create a simple calculator:
# Ask for two numbers and an operator (+, -, *, /)
# Use if statements to compute the result

operator = input("Enter an operator (+, -, *, /)")
# Addition
def Addition():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

#Subtraction
def Subtraction():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    if num1 > num2:
      result = num1 - num2
      print(f"{num1} - {num2} = {result}")
    else:
      print("First number needs to be greater!")

def Multiplication():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")

def Division():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    try:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

if operator == "+":
    Addition()
elif operator == "-":
    Subtraction()
elif operator == "*":
    Multiplication()
elif operator == "/":
    Division()
else:
    print("Invalid operator. Please enter +, -, *, or /.")
```

```python
# 11. Check if a year is a leap year:
# leap years are those that are divisible by 4 e.g. 2020
# leap years are not divisible by 100 except for those that are also divisible by 400

leap_yr = int(input("Enter a year to check if it is a leap year: "))
if (leap_yr % 4 == 0 and leap_yr % 100 != 0) or (leap_yr % 400 == 0):
  print(f"{leap_yr} is a leap year")
else:
  print(f"{leap_yr} is not a leap year")
```

```python
# 15. Ask the user for marks in three subjects. If the average is above 50, print "Passed"; otherwise, print "Failed".
marks =[]
sub1 = float(input("Enter marks for subject 1: "))
marks.append(sub1)
sub2 = float(input("Enter marks for subject 2: "))
marks.append(sub2)
sub3 = float(input("Enter marks for subject 3: "))
marks.append(sub3)
average = sum(marks)/len(marks)
if average > 50:
  average = round(average, 2)
  print(f"Average: {average}, Passed")
else:
  print(f"Average: {average}, Failed")
```

```python
# 17. Ask the user for a number and check if it's a prime number (Hint: A prime number has only two factors)
num = int(input("Enter a number to check if it's a prime number: "))

# define a flag variable
flag = False

if num == 0 or num == 1:
    print(num, "is not a prime number")
elif num > 1:
    # check for factors
    for i in range(2, num):
        if (num % i) == 0:
            # if factor is found, set flag to True
            flag = True
            # break out of loop
            break

    # check if flag is True
    if flag:
        print(num, "is not a prime number")
    else:
        print(num, "is a prime number")
```

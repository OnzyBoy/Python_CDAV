```python
# 2. Print even numbers between 1 and 20
# Output:
# 2, 4, 6, ..., 20
# 3. Print the square of numbers from 1 to 5
# Output:
# 1 4 9 16 25
import math
print("Even Numbers between 1-20\n")
for i in range(1,20):
  if i%2==0:
    print(i)

print("Square of numbers between 1-5\n")
for i in range(1,6):
  print(f"Square of {i}: {math.pow(i,2)}")
```

```python
# 7. Keep asking the user to enter a password until they type "secret"
# Output:
# Try again... until correct
while True:
  password = input("Enter Password: ")
  password = password.lower()
  if password == "secret":
    print("Correct")
    break
  print("Try again... until correct")
```

```python
# 8. Ask the user to enter numbers until they enter 0. Then print the sum.
# Input:
# 5
# 3
# 0
# Output:
# Total: 8
total = 0
while True:
  num = int(input("Enter Number: "))
  if num == 0:
    print("Getting Total... \n ")
    break
  total += num
print(f"Total: {total}")
```

```python
# 11. Count how many times "apple" appears in the list:
fruits = ["apple", "banana", "apple", "cherry", "apple"]
count = 0
for f in fruits:
  if f == "apple":
    count += 1
print(f"Apple Count: {count}")
```

```python
# 12. Use a loop to reverse a string:
# Input:
# "hello"
# Output:
# "olleh"

text = "Hello"
reversed_str = ""

for char in text:
    reversed_str = char + reversed_str

print(reversed_str)
```

```python
# 14. Use a loop to calculate the factorial of a number:
# Input:
# 5
# Output:
# 120

#With Math Import
# import math

# def factorial(n):
#   if n <0:
#     return  "Factorial for negative numbers is udefined"
#   else:
#     return math.factorial(n)
# num = int(input("Enter a number to find its factorial: "))
# print(f"Factorial of {num} is {factorial(num)}")

#For loop
num = int(input("Enter a number to find its factorial: "))
# Initialize the factorial variable to 1
fact = 1
# Calculate the factorial using a for loop
for i in range(1, num + 1):
  if num<0:
    print("Factorial for negative numbers is udefined")
  else:
    fact *= i

print(f"Factorial of {num} is {fact}")
```

```python
# 15. Loop from 1 to 100 and:
# Print "Hit" if divisible by 3
# Print "Miss" if divisible by 5
# Print "Jackpot" if divisible by both
for x in range(1,101):
  if x%3==0 and x%5==0:
    print(f"{x}: Jackpot")
  elif x%3==0:
    print(f"{x}: Hit")
  elif x%5==0:
    print(f"{x}: Miss")
```
